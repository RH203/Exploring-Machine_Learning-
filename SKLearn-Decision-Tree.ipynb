```python
import pandas as pd

# Read Iris sample
iris = pd.read_csv('sample/Iris.csv')
```

```python
'''
Untuk membaca informasi yang didapat setelah membaca file csv bisa menggunakan pandas dengan menggunakan
fungsi info() atau fungsi head() (Fungsi head hanya membaca lima data teratas pada dataset)
'''
iris.info()
```

```python
iris.head()
```

```python
'''
Bisa dilihat bahwa ada satu kolom yang tidak kita butuhkan yaitu kolom id, untuk menghilangkan
kolom id bisa menggunakan fungsi drop(). Parameter drop():
1. 'Id': Ini adalah label dari kolom yang ingin dihapus dari DataFrame iris. Dalam hal ini, kita menghapus kolom yang disebut 'Id'.
2. axis=1: Parameter ini menunjukkan bahwa yang ingin kita hapus adalah kolom ('Id'), bukan baris. Nilai axis=1 menunjukkan bahwa kita akan beroperasi pada sumbu kolom. Nilai axis=0 akan menunjukkan untuk menghapus baris.
3. inplace=True: Ketika inplace diatur sebagai True, itu berarti operasi ini akan mempengaruhi DataFrame iris itu sendiri, dan perubahan akan berlangsung di tempat. Artinya, DataFrame akan diubah tanpa perlu menetapkan hasilnya ke variabel baru. Jika inplace diatur sebagai False atau tidak disertakan, 
    maka itu akan mengembalikan DataFrame baru dengan perubahan yang diterapkan.
'''
iris.drop('Id', axis=1, inplace=True)
```

```python
iris.info()
```

```python
# Sebelum train model memisahan atribut dengan label
x = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=123)
```

```python
# Train model menggunakan fungsi fit()
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

tree_model = tree_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
 
y_pred = tree_model.predict(x_test)
 
acc_secore = round(accuracy_score(y_pred, y_test), 3)
 
print('Accuracy: ', acc_secore)
```

```python
# prediksi model dengan tree_model.predict([[SepalLength, SepalWidth, PetalLength, PetalWidth]])
print(tree_model.predict([[3, 3.4, 5.4, 2.3]])[0])
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file = "iris_tree.dot",
    feature_names = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],
    class_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ],
    rounded= True,
    filled =True)
```
