# Data Preparation with One-Hot-Encoding Technique

Teknik One-Hot-Encoding biasa digunakan kepada dataset yang tidak bisa dibandingkan ( Data Kategorik ) dan sebuah model machine learning umumnya tidak dapat mengolah data kategorik. Karena hal tersebut one hot encoding dapat diterapkan setelah data cleaning. Beberapa model seperti Regresi Linear dan Support Vector Machine hanya menerima input berupa numerik.

**Contoh sebuah kolom Data Kategorik:** 
- Kolom cuaca seperti cerah, berawan, dan hujan
- Kolom buah seperti apel, pisang , dan durian


```python
import pandas as pd
```

```python
df = pd.read_csv('sample/weather_one_hot_encoding_sample.csv')
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
# Data
cuaca = pd.read_csv('sample/weather_one_hot_encoding_sample.csv')
cuaca
```

```python
# Membuat instance OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
encoder
```

```python
cuaca_encoded = pd.DataFrame(encoder.fit_transform(np.array(cuaca).reshape(-1, 1)))
cuaca_encoded
```
